### –£–±—Ä–∞—Ç—å —Ä–µ–¥–∫–∏–µ –∏ –≤—ã–±–∏–≤–∞—é—â–∏–µ—Å—è –∑–Ω–∞—á–µ–Ω–∏—è. –û–ø–∏—Å–∞—Ç—å, –∫–∞–∫–∏–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã.

–£–¥–∞–ª—è—Ç—å –≤—ã–±—Ä–æ—Å—ã —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–æ –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –∫–≤–∞—Ä—Ç–∏–ª—è—Ö.

````python
# –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –æ—á–∏—Å—Ç–∫–∏ —Å—Ç–æ–ª–±—Ü–∞ –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤
def remove_fakes(column,data):
    q25=np.array(data[column].describe(percentiles=[.25,.75]).loc['25%'])
    q75=np.array(data[column].describe(percentiles=[.25,.75]).loc['75%'])
    first_part=q25-1.5*(q75-q25)
    second_part=q75+1.5*(q75-q25)
    count = 0
    leng = 1
    while leng > 0:
        index_del= []
        for index, element in enumerate(data[column]):
            if first_part > element or second_part< element:
                index_del.extend(data[data[column] == element].index)

        index_del= set(index_del)
        leng = len(index_del)
        count += 1
        print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫, –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è: {}'.format(leng))
        print(leng)
        data.drop(index_del,axis=0,inplace=True)
````
